**Задача 4.3.1** Задайте массив случайных значений из интервала(0; 1).
Рассчитайте средние и медианные значения для массива, сравните
результаты, какие выводы можно сделать о значениях?
Постройте точечную диаграмму рассения полученного ряда.

```python
import seaborn
from random import uniform


arr = []
cnt = 0

for i in range(10):
  n = uniform(0.0, 1.0)
  cnt += n
  arr.append(n)

ar = arr
ar.sort()
mid = len(ar)//2
res = (arr[mid] + arr[~mid])/2

print(f"Медиана: {res}")
print(f"Ср.значение: {cnt/10}")

seaborn.set(style='whitegrid')

seaborn.scatterplot(data=arr)

```

**Задание 4.3.4** Откройте статистические данные по стоимости акций Apple,

для двух переменных ′ + ′, ′ − ′, ′ ∙ ′, ′ ÷ ′, а также вычисление
следующих функций: 𝑒
𝑥+𝑦
, sin(𝑥 + 𝑦) , cos(𝑥 + 𝑦) , 𝑥
𝑦
.

```python
import numpy as np


def functions():
  print('------ФУНКЦИИ------\n|__ 1.exp\n|\n|__ 2.sin\n|\n|__ 3.cos\n|\n|__ 4.x^y')

  n = int(input('Введите от 1 до 4, какую операцию вы хотите выполнить: '))

  x = float(input('x= '))
  y = float(input('y= '))

  if n == 1:
     ans = np.exp(x + y)
  elif n == 2:
    ans = np.sin(x + y)
  elif n == 3:
    ans = np.cos(x + y)
  elif n == 4:
    ans = x**y
  else:
    print('Error')

  print(f'Ответ: {ans}')


def calculator():
  print('------ПРОСТЕЙШИЙ КАЛЬКУЛЯТОР------\n|__ +\n|\n|__ -\n|\n|__ *\n|\n|__ /')
  n = input('Введите знак действия: ')

  x = float(input('x= '))
  y = float(input('y= '))

  if n == '+':
    ans = x + y
  elif n == '-':
    ans = x - y
  elif n == '*':
    ans = x * y
  elif n == '/':
    ans = x / y
  else:
    print('Error')

  print("Ответ: {ans}")


if __name__ == '__main__':
  print('------КАЛЬКУЛЯТОР------')
  print("Введите что выхоите вычислить:\n\t|__ 1 - если хотите запустить"
  "простейший калькулятор\n\t|__ 2 - если хотите вычислить функцию")
  a = int(input())

  if a == 1:
    calculator()
  elif a == 2:
    functions()
  else:
    print('Error')
```
