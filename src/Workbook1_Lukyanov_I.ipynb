–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞.

```python
import seaborn
from random import uniform


arr = []
cnt = 0

for i in range(10):
  n = uniform(0.0, 1.0)
  cnt += n
  arr.append(n)

ar = arr
ar.sort()
mid = len(ar)//2
res = (arr[mid] + arr[~mid])/2

print(f"–ú–µ–¥–∏–∞–Ω–∞: {res}")
print(f"–°—Ä.–∑–Ω–∞—á–µ–Ω–∏–µ: {cnt/10}")

seaborn.set(style='whitegrid')

seaborn.scatterplot(data=arr)

```

–¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤ + ‚Ä≤, ‚Ä≤ ‚àí ‚Ä≤, ‚Ä≤ ‚àô ‚Ä≤, ‚Ä≤ √∑ ‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ

, sin(ùë• + ùë¶) , cos(ùë• + ùë¶) , ùë•

```python
import numpy as np


def functions():
  print('------–§–£–ù–ö–¶–ò–ò------\n|__ 1.exp\n|\n|__ 2.sin\n|\n|__ 3.cos\n|\n|__ 4.x^y')

  n = int(input('–í–≤–µ–¥–∏—Ç–µ –æ—Ç 1 –¥–æ 4, –∫–∞–∫—É—é –æ–ø–µ—Ä–∞—Ü–∏—é –≤—ã —Ö–æ—Ç–∏—Ç–µ –≤—ã–ø–æ–ª–Ω–∏—Ç—å: '))

  x = float(input('x= '))
  y = float(input('y= '))

  if n == 1:
     ans = np.exp(x + y)
  elif n == 2:
    ans = np.sin(x + y)
  elif n == 3:
    ans = np.cos(x + y)
  elif n == 4:
    ans = x**y
  else:
    print('Error')

  print(f'–û—Ç–≤–µ—Ç: {ans}')


def calculator():
  print('------–ü–†–û–°–¢–ï–ô–®–ò–ô –ö–ê–õ–¨–ö–£–õ–Ø–¢–û–†------\n|__ +\n|\n|__ -\n|\n|__ *\n|\n|__ /')
  n = input('–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞–∫ –¥–µ–π—Å—Ç–≤–∏—è: ')

  x = float(input('x= '))
  y = float(input('y= '))

  if n == '+':
    ans = x + y
  elif n == '-':
    ans = x - y
  elif n == '*':
    ans = x * y
  elif n == '/':
    ans = x / y
  else:
    print('Error')

  print("–û—Ç–≤–µ—Ç: {ans}")


if __name__ == '__main__':
  print('------–ö–ê–õ–¨–ö–£–õ–Ø–¢–û–†------')
  print("–í–≤–µ–¥–∏—Ç–µ —á—Ç–æ –≤—ã—Ö–æ–∏—Ç–µ –≤—ã—á–∏—Å–ª–∏—Ç—å:\n\t|__ 1 - –µ—Å–ª–∏ —Ö–æ—Ç–∏—Ç–µ –∑–∞–ø—É—Å—Ç–∏—Ç—å"
  "–ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä\n\t|__ 2 - –µ—Å–ª–∏ —Ö–æ—Ç–∏—Ç–µ –≤—ã—á–∏—Å–ª–∏—Ç—å —Ñ—É–Ω–∫—Ü–∏—é")
  a = int(input())

  if a == 1:
    calculator()
  elif a == 2:
    functions()
  else:
    print('Error')
```
