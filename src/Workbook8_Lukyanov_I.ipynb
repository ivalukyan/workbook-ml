**Пример 1.1.2**
Проведем кластеризацию К-средних к набору простых цифр. К-means попытается индентифицировать похожие цифры

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np

from sklearn.cluster import KMeans
from sklearn.datasets import load_digits


digits = load_digits()
digits.data.shape
kmeans = KMeans(n_clusters= 10, random_state=0)
clusters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape

fig, ax = plt.subplots(2, 5, figsize = (8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8 ,8)

for axi, center in zip(ax.flat, centers):
    axi.set(xticks =[], yticks = [])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

**Задание 1.1.3**

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


X = np.array([[5,3], [10,15], [15,12], [24,10], [30,45], [85,70], [71,80], [60,78], [55,52], [80,91]])


def kmeans_clustering(k):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    
    labels = kmeans.labels_
    
    centroids = kmeans.cluster_centers_

    plt.figure(figsize=(8, 6))
    for i in range(k):
        cluster_points = X[labels == i]
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i+1}')
    plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='x', label='Centroids')
    plt.title(f'KMeans Clustering with k={k}')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    plt.show()

for k in range(2, 6):
    kmeans_clustering(k)
```

**Задание 1.1.4**

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np


iris = load_iris()
X = iris.data
y = iris.target


def kmeans_clustering(X, k):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    return kmeans.labels_


for k in range(2, 6):
    labels = kmeans_clustering(X, k)
    cluster_df = pd.DataFrame(np.column_stack((iris.target, labels)), columns=['Actual Class', 'Predicted Cluster'])
    print(f"Results for k={k}:")
    print(cluster_df)
    print()
```
